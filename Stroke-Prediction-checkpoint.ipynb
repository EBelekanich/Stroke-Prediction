```python
# drop the lifestyle columns from the dataset
stroke_health= stroke_df.drop(columns =['Residence_type', 'work_type', 'ever_married','id'])
stroke_health.head()
```

```python
#feature engineering Create new column for blood sugar with 4 categories
stroke_health['Blood_sugar'] = pd.cut(stroke_health.avg_glucose_level, bins=4,
       labels=["low", "normal", "borderline","high"]).head()
stroke_health.head()
```

```python
#partion the dataset into 60% validation and 40% train split

trainData, validData = train_test_split(stroke_health, test_size=0.4, random_state=2) # partition the dataset
print( trainData.shape, validData.shape)

#split the dataset predictor and outcome variables
predictors = list(stroke_health.columns) # predictor variables
outcome = 'stroke'# outcome variable
predictors.remove(outcome)
predictors.remove("avg_glucose_level") # drop the avg_glucose
print(predictors)
train_X = trainData[predictors]
train_y = trainData[outcome]
valid_X = validData[predictors]
valid_y = validData[outcome]
```
